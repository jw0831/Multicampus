```
from google.colab import files
import io
from konlpy.tag import Okt
from konlpy.tag import Kkma
import json
import os
import pandas as pd
import re
```

```
# 파일 불러오기
myfile = files.upload()
```

```
jeju_blog = pd.read_csv(io.BytesIO(myfile['jeju_place.csv']))
jeju_blog.head()
```

```
kkma = Kkma()
okt=Okt()
```

```
pd.Series(jeju_blog['description'])
```

```
review_text = ''
for line in jeju_blog['description'][0:1300]:
    review_text = review_text + line
    review = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',review_text)
```

```
review
```

```
# 형태소 분석
tokens = okt.morphs(review, stem = True)
tokens
```

```
from collections import Counter
```

```
count = Counter(tokens)
```

```
# 많이나온 형태소 개수
count.most_common()
```

```
# 명사분석
word = kkma.nouns(review)
word
```

```
word_count = Counter(word)
word_count.most_common()
```
