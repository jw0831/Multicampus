```
okt=Okt()
kkma = Kkma()
```

```
len(blog_kakao)
```

```
blog_kakao['contents'] = blog_kakao['contents'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```
blog_kakao['contents'] = blog_kakao['contents'].astype(str)
```

```
kakao_tokens = []
for sentence in blog_kakao['contents']:
    kakao_review = okt.morphs(sentence, stem=True) 
    kakao_tokens.append(kakao_review)
```

```
blog_naver['description'] = blog_naver['description'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```
blog_naver['description'] = blog_naver['description'].astype(str)
```

```
naver_tokens = []
for sentence in blog_naver['description'] :
    naver_review = okt.morphs(sentence, stem=True) 
    naver_tokens.append(naver_review)
```

```
review_text = ''
for line in mango['댓글']:
    review_text = review_text + line
    mango_review = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',review_text)
```

```
mango_review
```

```
mango_tokens = okt.morphs(mango_review, stem = True)
mango_tokens
```

```
review_text = ''
for line in dining['댓글']:
    review_text = review_text + line
    review_dining = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',review_text)
```

```
review_dining
```

```
dining_tokens = okt.morphs(review_dining, stem = True)
dining_tokens
```

```
review_text = ''
for line in insta['narration']:
    review_text = review_text + line
    review_insta = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',review_text)
```

```
review_insta
```

```
insta_tokens = okt.morphs(review_insta, stem = True)
insta_tokens
```

```
# 블로그 형태소는 2차원 배열이라 1차원 배열로 바꾸어줌
import itertools
kakao_tokens = list(itertools.chain(*kakao_tokens))
```

```
naver_tokens = list(itertools.chain(*naver_tokens))
```

```
# 각 리뷰, 블로그 글내용에서 불용어 제거(찬주님이 함)
```

```
file_stopwords = '/content/drive/My Drive/프로젝트1/stopwords_KOR.csv'
stopwords = pd.read_csv(file_stopwords,encoding='utf-8')
stopwords.head()
```

```
# 불용어 사전 리스트
stop_list=list(stopwords.iloc[:,0].values)
len(stop_list) # 732개
```

```
# 형태소 파일 읽어오기
word_tokens=list(token_counts_totaldf['형태소'])
print('불용어 제거 전: ', len(word_tokens))
word_tokens=list(token_counts_totaldf['형태소'])
token_counts_totaldf
```

```
# 불용어 사전에 포함되어 있는 형태소의 인덱스 리스트 생성
idx_list=[]
for i, w in enumerate(word_tokens):
    if w in stop_list:
        idx_list.append(i)
```

```
after=token_counts_totaldf.drop(idx_list) # 45672개
after.to_csv('/content/drive/My Drive/프로젝트1/total_tokens_count2_nostopwords.csv',encoding='utf-8',index=False)
```
