```python
import pandas as pd
import numpy as np
```

```python
mango = pd.read_excel("mango_crawling_df.xlsx")
mango = mango.drop('Unnamed: 0',axis=1)
mango
```

```python
dining = pd.read_excel("code_crawling_df.xlsx")
dining = dining.drop("Unnamed: 0",axis=1)
dining
```

```python
name_dining=dining['가게명']
name_mango=mango['가게명']
name_mango = list(name_mango)
name_dining = list(name_dining)
```

```python
name_dining_list=[]
for i in name_dining:
    name_dining_list.append([i])
name_dining_list
```

```python
name_mango_list=[]
for i in name_mango:
    name_mango_list.append([i])
name_mango_list
```

```python
temp_not_in=[]
for i in name_mango_list:
    if i not in name_dining_list:
        temp_not_in.append(i)
print(len(temp_not_in))
temp_not_in
```

```python
temp_in=[]
for i in name_mango_list:
    if i in name_dining_list:
        temp_in.append(i)
print(len(temp_in))
temp_in
```
