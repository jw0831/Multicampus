```python
import pandas as pd
import numpy as np
```

```python
import re
```

```python
df=pd.read_csv("review_totals.csv")
df
```

```python
def clean_reviews(reviews):
    temp=reviews.split('\\n')
    r_list=[]
    for t in temp:
        text = re.sub('^[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', t.strip())
        if len(text)>1:
            if re.match('^[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', text[0]):
                continue
            else:
                res="'"+text+"'"
                r_list.append(res)
    return r_list
```

```python
def list_to_string(review_list):
    result=""
    if len(review_list)>0:
        result = ', '.join(review_list)
    return result
```

```python
def clean_dining(reviews):
    return reviews[1:-1]
```

```python
df['리뷰_x']=df['리뷰_x'].fillna("").apply(lambda x: clean_reviews(x)).apply(lambda x: list_to_string(x))
df['리뷰_y']=df['리뷰_y'].fillna("").apply(lambda x: clean_dining(x))
df['리뷰']=df['리뷰'].fillna("").apply(lambda x: clean_reviews(x)).apply(lambda x: list_to_string(x))
df['total']=df['리뷰_x']+df['리뷰_y']+df['리뷰']
```
