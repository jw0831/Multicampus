# Train, Test, Validation 폴더 정리

```python
testlist= os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/test")
for j in range(len(testlist)):
    filenames = os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/test/"+testlist[j])
    removefilename = [ i for i in filenames if i.startswith('.')]
    for i in range(len(removefilename)):
        try:
            os.remove("C:/Users/USER/Desktop/여기요/서울시_ttv/test/"+testlist[j]+"/"+removefilename[i])
        except Exception:
            print(Exception)
```

```python
testlist= os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/train")
for j in range(len(testlist)):
    filenames = os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/train/"+testlist[j])
    removefilename = [ i for i in filenames if i.startswith('.')]
    for i in range(len(removefilename)):
        try:
            os.remove("C:/Users/USER/Desktop/여기요/서울시_ttv/train/"+testlist[j]+"/"+removefilename[i])
        except Exception:
            print(Exception)
```

```python
testlist= os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/val")
for j in range(len(testlist)):
    filenames = os.listdir("C:/Users/USER/Desktop/여기요/서울시_ttv/val/"+testlist[j])
    removefilename = [ i for i in filenames if i.startswith('.')]
    for i in range(len(removefilename)):
        try:
            os.remove("C:/Users/USER/Desktop/여기요/서울시_ttv/val/"+testlist[j]+"/"+removefilename[i])
        except Exception:
            print(Exception)
```
